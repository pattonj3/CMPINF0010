```python
import time

name = input("Please state your name: ")
age = 0
while True:
    try:
        age = int(input("Please enter your age: "))
        break
    except:
        continue
print("Name: " + name + "\n" + "Age: " + str(age))

# information sourced from https://iard.org/science-resources/detail/Minimum-Legal-Age-Limits
# Each number represents how many countries have that age as the minimum drinking age.
# We start with 8 countries which have no minimum age, then add for each age.
numDrinks = 8
if age >= 14:
    numDrinks += 1
if age >= 16:
    numDrinks += 16
if age >= 17:
    numDrinks += 1
if age >= 18:
    numDrinks += 127
if age >= 19:
    numDrinks += 1
if age >= 20:
    numDrinks += 6
if age >= 21:
    numDrinks += 19

print("Hey " + name + "! You are old enough to purchase alcohol in " + str(numDrinks) + " different countries. You know what that means...")
print("It's time for a road trip 'round the world!\n")
while (numDrinks > 0):
    print(str(numDrinks) + " bottles of beer on the wall,")
    print(str(numDrinks) + " bottles of beer!")
    print("Take one down, pass it around,")
    numDrinks -= 1
    print(str(numDrinks) + " bottles of beer on the wall.\n")
    time.sleep(0.5)
print("Are we there yet? I'm feeling a little woozy...")
```
